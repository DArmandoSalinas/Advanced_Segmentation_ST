```python
# --- Imports & notebook display setup
import pandas as pd
import numpy as np
import warnings
from datetime import datetime
from collections import Counter
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
from matplotlib.patches import FancyBboxPatch, FancyArrowPatch

warnings.filterwarnings("ignore")

pd.set_option("display.max_columns", 120)
pd.set_option("display.width", 140)
```

```python
# ============================================================================
# Historical Data Parsing & APREU Activity Classification
# ============================================================================

# --- Helper functions for parsing HubSpot historical data (// delimited)
def hist_all(val):
    """Parse HubSpot history string and return ALL values as a list."""
    if pd.isna(val):
        return []
    s = str(val).strip()
    if "//" in s:
        parts = [p.strip() for p in s.split("//") if p.strip() != ""]
        return [p for p in parts if p.lower() not in ['nan', 'none', '']]
    return [s] if s.lower() not in ['nan', 'none', ''] else []

def hist_latest(val):
    """Return the latest (last) value from HubSpot history string."""
    if pd.isna(val):
        return np.nan
    s = str(val).strip()
    if "//" in s:
        parts = [p.strip() for p in s.split("//") if p.strip() != ""]
        if not parts:
            return np.nan
        return parts[-1]
    return s

def hist_concat_text(val):
    """Concatenate all historical text values with space separator."""
    values = hist_all(val)
    return " ".join(values)

# --- APREU Activity Classification Dictionaries
# Based on common APREU promotional activities seen in the data

DIGITAL_ACTIVITIES = [
    'sitio web', 'sitio', 'website', 'web', 
    'formulario', 'formulario rua', 'rua', 'form',
    'google ads', 'facebook ads', 'ads', 'paid search',
    'organic search', 'organic', 'seo',
    'landing page', 'lp'
]

EVENT_ACTIVITIES = [
    'open day', 'openday', 'open house', 
    'fogatada', 'fogata',
    'tdla', 'tour de la admision', 'tour admisi√≥n',
    'gira panama', 'gira panam√°', 'panama', 'panam√°',
    'feria', 'feria universitaria', 'expo',
    'evento carrera', 'eventos carreras',
    'visita campus', 'campus tour', 'recorrido',
    'conferencia', 'charla', 'webinar',
    'd√≠a de puertas abiertas'
]

MESSAGING_ACTIVITIES = [
    'whatsapp', 'whats app', 'wa', 
    'mensaje directo', 'direct message', 'dm',
    'chat', 'messenger',
    'contacto directo', 'direct contact',
    'llamada', 'phone call', 'call'
]

NICHE_ACTIVITIES = [
    'lion leaders', 'lion leader', 'leaders',
    'programa especial', 'special program',
    'beca', 'scholarship', 'becas',
    'intercambio', 'exchange',
    'embajador', 'ambassador', 'embajadores',
    'referido', 'referral', 'referred',
    'alumni', 'ex-alumno'
]

# Function to detect activity type from text
def detect_activity_type(text, activity_dict):
    """
    Search for activity keywords in text and return count of matches.
    """
    if pd.isna(text) or text == "":
        return 0
    
    text_lower = str(text).lower()
    count = 0
    
    for keyword in activity_dict:
        if keyword in text_lower:
            count += 1
    
    return count

# Function to classify entry channel based on activities
def classify_entry_channel(apreu_hist, first_conv, recent_conv):
    """
    Classify contact into entry channel segment (3A/3B/3C/3D) based on:
    - Historical APREU activities
    - First conversion event
    - Recent conversion event
    
    Priority order:
    1. Check APREU activities for clear signals
    2. Check first conversion event
    3. Default to "Unknown" if no clear signal
    """
    # Combine all text for analysis
    all_text = f"{apreu_hist} {first_conv} {recent_conv}".lower()
    
    # Count each type of activity
    digital_count = detect_activity_type(all_text, DIGITAL_ACTIVITIES)
    event_count = detect_activity_type(all_text, EVENT_ACTIVITIES)
    messaging_count = detect_activity_type(all_text, MESSAGING_ACTIVITIES)
    niche_count = detect_activity_type(all_text, NICHE_ACTIVITIES)
    
    # Find dominant type (with tie-breaking)
    activity_scores = {
        '3A_Digital': digital_count,
        '3B_Event': event_count,
        '3C_Messaging': messaging_count,
        '3D_Niche': niche_count
    }
    
    # Get max score
    max_score = max(activity_scores.values())
    
    # If no signals found, return Unknown
    if max_score == 0:
        return 'Unknown'
    
    # Return the category with highest score
    # If tie, priority: Event > Digital > Messaging > Niche
    if activity_scores['3B_Event'] == max_score and max_score > 0:
        return '3B_Event'
    elif activity_scores['3A_Digital'] == max_score and max_score > 0:
        return '3A_Digital'
    elif activity_scores['3C_Messaging'] == max_score and max_score > 0:
        return '3C_Messaging'
    elif activity_scores['3D_Niche'] == max_score and max_score > 0:
        return '3D_Niche'
    else:
        return 'Unknown'

print("‚úÖ Historical parsing functions and APREU activity classification ready")
```

```python
# ============================================================================
# Date Conversions & Time-to-Close Calculations
# ============================================================================

def convert_hubspot_timestamp(val):
    """Convert HubSpot timestamp (milliseconds since epoch) to datetime."""
    if pd.isna(val):
        return pd.NaT
    try:
        timestamp_ms = int(float(str(val).strip()))
        return pd.to_datetime(timestamp_ms, unit='ms')
    except (ValueError, TypeError):
        return pd.NaT

# Convert date fields
date_fields = ["create_date", "close_date", "first_conversion_date", "recent_conversion_date"]

for field in date_fields:
    if field in X.columns:
        print(f"Converting {field} from HubSpot timestamps...")
        X[field] = X[field].apply(convert_hubspot_timestamp)
        non_null_count = X[field].count()
        print(f"  ‚úÖ Successfully converted {non_null_count:,} timestamps")

# Calculate days_to_close
if "create_date" in X.columns and "close_date" in X.columns:
    print("\nüìä Calculating days_to_close...")
    raw_days = (X["close_date"] - X["create_date"]).dt.days
    X["days_to_close"] = raw_days.where(raw_days >= 0)
    
    negative_count = (raw_days < 0).sum()
    if negative_count > 0:
        print(f"  ‚ö†Ô∏è Filtered out {negative_count:,} contacts with negative days (data quality)")
    
    # Create time-to-close buckets
    def categorize_ttc(days):
        if pd.isna(days):
            return "Still Open"
        elif days <= 30:
            return "Early (‚â§30 days)"
        elif days <= 60:
            return "Medium (31-60 days)"
        elif days <= 120:
            return "Late (61-120 days)"
        else:
            return "Very Late (>120 days)"
    
    X["ttc_bucket"] = X["days_to_close"].apply(categorize_ttc)
    dtc_valid = X["days_to_close"].notna().sum()
    print(f"  ‚úÖ Calculated days_to_close for {dtc_valid:,} closed contacts")
else:
    X["days_to_close"] = np.nan
    X["ttc_bucket"] = "Unknown"

# Calculate conversion journey duration (first to recent conversion)
if "first_conversion_date" in X.columns and "recent_conversion_date" in X.columns:
    print("\nüìä Calculating conversion journey duration...")
    raw_journey = (X["recent_conversion_date"] - X["first_conversion_date"]).dt.days
    X["conversion_journey_days"] = raw_journey.where(raw_journey >= 0)
    journey_valid = X["conversion_journey_days"].notna().sum()
    print(f"  ‚úÖ Calculated journey duration for {journey_valid:,} contacts")
else:
    X["conversion_journey_days"] = np.nan

print("\n‚úÖ Date conversions and time calculations complete")
```

```python
# ============================================================================
# Preparatoria Consolidation & Normalization
# ============================================================================

print("üè´ Consolidating preparatoria data...")

# Consolidate preparatoria from multiple fields
def consolidate_prepa(row):
    """Consolidate preparatoria from multiple possible fields."""
    for field in ["prep_bpm", "prep_name", "prep_donde_estudia"]:
        if field in row and pd.notna(row[field]) and str(row[field]).strip() not in ['', 'nan', 'None']:
            return str(row[field]).strip()
    return "Unknown"

X["preparatoria"] = X.apply(consolidate_prepa, axis=1)

# Normalize preparatoria year
if "prep_year" in X.columns:
    def normalize_year(val):
        if pd.isna(val):
            return "Unknown"
        s = str(val).strip().lower()
        if any(x in s for x in ['1', 'primer', 'first', 'uno']):
            return "1st Year"
        elif any(x in s for x in ['2', 'segundo', 'second', 'dos']):
            return "2nd Year"
        elif any(x in s for x in ['3', 'tercer', 'third', 'tres']):
            return "3rd Year"
        else:
            return "Unknown"
    
    X["prep_year_normalized"] = X["prep_year"].apply(normalize_year)
else:
    X["prep_year_normalized"] = "Unknown"

prepa_count = (X["preparatoria"] != "Unknown").sum()
print(f"  ‚úÖ Consolidated preparatoria data for {prepa_count:,} contacts")
print(f"  üìä Unique preparatorias: {X['preparatoria'].nunique()}")
print(f"\n Preparatorias:")
print(X["preparatoria"].value_counts())

print("\n‚úÖ Preparatoria consolidation complete")
```

```python
# ============================================================================
# Feature Engineering - Engagement & Email Metrics
# ============================================================================

print("‚öôÔ∏è Engineering features...")

# Convert numeric fields to float
numeric_fields = ["num_sessions", "num_pageviews", "forms_submitted", 
                 "email_delivered", "email_opened", "email_clicked", "likelihood_to_close"]

for field in numeric_fields:
    if field in X.columns:
        X[field] = pd.to_numeric(X[field], errors='coerce').fillna(0)

# Log-transformed engagement metrics (add 1 to avoid log(0))
X["log_sessions"] = np.log1p(X["num_sessions"])
X["log_pageviews"] = np.log1p(X["num_pageviews"])
X["log_forms"] = np.log1p(X["forms_submitted"])

# Engagement score (combined metric)
X["engagement_score"] = X["log_sessions"] + X["log_pageviews"] + X["log_forms"]

# Efficiency ratios
X["pageviews_per_session"] = np.where(
    X["num_sessions"] > 0, 
    X["num_pageviews"] / X["num_sessions"], 
    0
)

X["forms_per_session"] = np.where(
    X["num_sessions"] > 0,
    X["forms_submitted"] / X["num_sessions"],
    0
)

# Email engagement score
email_open_rate = np.where(
    X["email_delivered"] > 0,
    X["email_opened"] / X["email_delivered"],
    0
)

email_click_rate = np.where(
    X["email_opened"] > 0,
    X["email_clicked"] / X["email_opened"],
    0
)

X["email_engagement_score"] = (email_open_rate * 0.5) + (email_click_rate * 0.5)

# Normalize likelihood to close to 0-100 scale
if "likelihood_to_close" in X.columns:
    # Handle both 0-1 and 0-100 scales
    max_val = X["likelihood_to_close"].max()
    if max_val > 1:
        X["likelihood_pct"] = X["likelihood_to_close"]
    else:
        X["likelihood_pct"] = X["likelihood_to_close"] * 100
else:
    X["likelihood_pct"] = 0

# Create binary close flag
X["is_closed"] = X["close_date"].notna().astype(int)

print(f"  ‚úÖ Engineered engagement features") 
print(f"  üìä Avg engagement score: {X['engagement_score'].mean():.2f}")
print(f"  üìä Avg email engagement: {X['email_engagement_score'].mean():.2%}")
print(f"  üìä Overall close rate: {X['is_closed'].mean():.2%}")

print("\n‚úÖ Feature engineering complete")
```

```python
# ============================================================================
# APREU Journey Visualization
# ============================================================================

def visualize_apreu_journey(contact_id):
    """
    Visualize the APREU activities journey of a contact over time.
    Shows APREU activities on the top line and conversions on the bottom line.
    """
    # Find the contact
    key = str(contact_id).strip()
    idx = X[X["contact_id"].astype(str) == key].index
    if len(idx) == 0:
        print(f"‚ùå No contact found with contact_id = {contact_id}")
        return
    
    contact_idx = idx[0]
    
    # Get APREU activities list
    activities = X.loc[contact_idx, "apreu_activities_list"]
    first_conv = X.loc[contact_idx, "first_conversion"]
    recent_conv = X.loc[contact_idx, "recent_conversion"]
    segment = X.loc[contact_idx, "segment_c3"]
    
    # Separate activities and conversions
    activity_steps = []
    conversion_steps = []
    
    # Add APREU activities
    if isinstance(activities, list) and activities and len(activities) > 0:
        for i, activity in enumerate(activities):
            activity_steps.append({
                'step': f'Activity {i+1}',
                'value': activity,
                'type': 'activity',
                'color': '#2196F3'  # Blue for activities
            })
    
    # Add first conversion if available
    if pd.notna(first_conv) and str(first_conv) not in ['', 'nan', 'None']:
        conversion_steps.append({
            'step': 'First Conversion',
            'value': first_conv,
            'type': 'conversion',
            'color': '#4CAF50'  # Green for conversions
        })
    
    # Add recent conversion if different from first
    if pd.notna(recent_conv) and str(recent_conv) not in ['', 'nan', 'None']:
        if recent_conv != first_conv:
            conversion_steps.append({
                'step': 'Recent Conversion',
                'value': recent_conv,
                'type': 'conversion',
                'color': '#FF9800'  # Orange for recent conversion
            })
    
    # If no data found
    if not activity_steps and not conversion_steps:
        print(f"‚ÑπÔ∏è No APREU journey data found for contact {contact_id}")
        return
    
    # Create visualization - TWO ROWS LAYOUT
    box_width = 2.5
    box_spacing = 0.8
    
    # Calculate width based on maximum items in either row
    max_items = max(len(activity_steps), len(conversion_steps))
    total_width = max(max_items * (box_width + box_spacing) + 1, 16)
    
    fig, ax = plt.subplots(figsize=(total_width, 9))
    ax.set_xlim(0, total_width)
    ax.set_ylim(0, 6.5)
    ax.axis('off')
    
    # Title
    ax.text(total_width / 2, 6.0, f'APREU Journey: Contact {contact_id}',
            ha='center', va='top', fontsize=16, fontweight='bold')
    ax.text(total_width / 2, 5.6, f'Segment: {segment}',
            ha='center', va='top', fontsize=12, style='italic')
    
    # Draw ACTIVITIES on TOP ROW
    if activity_steps:
        ax.text(0.5, 4.5, 'APREU Activities:',
                ha='left', va='bottom', fontsize=11, fontweight='bold',
                color='#2196F3')
        
        y_pos = 3.7
        current_x = 0.5
        
        for i, step in enumerate(activity_steps):
            # Draw box
            box = FancyBboxPatch(
                (current_x, y_pos - 0.4),
                box_width, 0.8,
                boxstyle="round,pad=0.1",
                facecolor=step['color'],
                edgecolor='darkgray',
                alpha=0.7,
                linewidth=2
            )
            ax.add_patch(box)
            
            # Add step label (above box)
            ax.text(current_x + box_width/2, y_pos + 0.6,
                    step['step'],
                    ha='center', va='bottom',
                    fontsize=9, fontweight='bold')
            
            # Add value text (inside box, wrapped if needed)
            value_text = step['value']
            if len(value_text) > 25:
                # Wrap long text
                words = value_text.split()
                lines = []
                current_line = []
                for word in words:
                    test_line = ' '.join(current_line + [word])
                    if len(test_line) <= 25:
                        current_line.append(word)
                    else:
                        if current_line:
                            lines.append(' '.join(current_line))
                        current_line = [word]
                if current_line:
                    lines.append(' '.join(current_line))
                value_text = '\n'.join(lines[:3])  # Max 3 lines
            
            ax.text(current_x + box_width/2, y_pos,
                    value_text,
                    ha='center', va='center',
                    fontsize=8, color='white', fontweight='bold',
                    wrap=True)
            
            # Draw arrow to next step (if not last)
            if i < len(activity_steps) - 1:
                arrow = FancyArrowPatch(
                    (current_x + box_width, y_pos),
                    (current_x + box_width + box_spacing, y_pos),
                    arrowstyle='->,head_width=0.4,head_length=0.4',
                    color='gray',
                    linewidth=2,
                    alpha=0.6
                )
                ax.add_patch(arrow)
            
            current_x += box_width + box_spacing
    
    # Draw CONVERSIONS on BOTTOM ROW
    if conversion_steps:
        ax.text(0.5, 2.3, 'Conversions:',
                ha='left', va='bottom', fontsize=11, fontweight='bold',
                color='#4CAF50')
        
        y_pos = 1.5
        current_x = 0.5
        
        for i, step in enumerate(conversion_steps):
            # Draw box
            box = FancyBboxPatch(
                (current_x, y_pos - 0.4),
                box_width, 0.8,
                boxstyle="round,pad=0.1",
                facecolor=step['color'],
                edgecolor='darkgray',
                alpha=0.7,
                linewidth=2
            )
            ax.add_patch(box)
            
            # Add step label (above box)
            ax.text(current_x + box_width/2, y_pos + 0.6,
                    step['step'],
                    ha='center', va='bottom',
                    fontsize=9, fontweight='bold')
            
            # Add value text (inside box, wrapped if needed)
            value_text = step['value']
            if len(value_text) > 25:
                # Wrap long text
                words = value_text.split()
                lines = []
                current_line = []
                for word in words:
                    test_line = ' '.join(current_line + [word])
                    if len(test_line) <= 25:
                        current_line.append(word)
                    else:
                        if current_line:
                            lines.append(' '.join(current_line))
                        current_line = [word]
                if current_line:
                    lines.append(' '.join(current_line))
                value_text = '\n'.join(lines[:3])  # Max 3 lines
            
            ax.text(current_x + box_width/2, y_pos,
                    value_text,
                    ha='center', va='center',
                    fontsize=8, color='white', fontweight='bold',
                    wrap=True)
            
            # Draw arrow to next step (if not last)
            if i < len(conversion_steps) - 1:
                arrow = FancyArrowPatch(
                    (current_x + box_width, y_pos),
                    (current_x + box_width + box_spacing, y_pos),
                    arrowstyle='->,head_width=0.4,head_length=0.4',
                    color='gray',
                    linewidth=2,
                    alpha=0.6
                )
                ax.add_patch(arrow)
            
            current_x += box_width + box_spacing
    
    # Add legend at the bottom with proper spacing
    activity_patch = mpatches.Patch(color='#2196F3', label='APREU Activity', alpha=0.7)
    conversion_patch = mpatches.Patch(color='#4CAF50', label='First Conversion', alpha=0.7)
    recent_patch = mpatches.Patch(color='#FF9800', label='Recent Conversion', alpha=0.7)
    ax.legend(handles=[activity_patch, conversion_patch, recent_patch],
             loc='lower center', bbox_to_anchor=(0.5, 0.02), ncol=3, frameon=False, fontsize=10)
    
    plt.tight_layout()
    plt.show()

print("‚úÖ visualize_apreu_journey function ready")

```

```python
# ============================================================================
# Activity Participation by Preparatoria - Detailed Percentage Analysis
# ============================================================================

print("üéØ Analyzing activity participation patterns by preparatoria...")

# Initialize empty dataframes in case of errors
prepa_activity_pct_df = pd.DataFrame()
prepa_activity_type_df = pd.DataFrame()

# Check if we have the required data
if len(top_prepa_list) == 0 or len(X_with_prepa) == 0:
    print("  ‚ö†Ô∏è No preparatoria data available for activity analysis")
else:
    # Get all unique activities from the top activities list
    top_activities_list = activity_df.head(15)["activity"].tolist() if len(activity_df) > 0 else []
    
    if len(top_activities_list) == 0:
        print("  ‚ö†Ô∏è No activities found for analysis")
    else:
        # Create activity participation matrix by preparatoria
        prepa_activity_participation = []
        
        for prepa in top_prepa_list[:20]:  # Top 20 prepas
            prepa_contacts = X_with_prepa[X_with_prepa["preparatoria"] == prepa]
            total_prepa_contacts = len(prepa_contacts)
            
            row_data = {
                "preparatoria": prepa,
                "total_contacts": total_prepa_contacts
            }
            
            # For each activity, calculate what % of this prepa's contacts attended it
            for activity in top_activities_list:
                contacts_with_activity = prepa_contacts[
                    prepa_contacts["apreu_activities_list"].apply(
                        lambda x: activity in x if isinstance(x, list) else False
                    )
                ]
                
                participation_pct = (len(contacts_with_activity) / total_prepa_contacts * 100) if total_prepa_contacts > 0 else 0
                row_data[activity] = round(participation_pct, 1)
            
            prepa_activity_participation.append(row_data)
        
        # Create DataFrame
        prepa_activity_pct_df = pd.DataFrame(prepa_activity_participation)
        
        # Sort by total contacts (descending)
        prepa_activity_pct_df = prepa_activity_pct_df.sort_values("total_contacts", ascending=False)
        
        print(f"\n  üìä Activity Participation % by Preparatoria")
        print(f"  (Shows what % of each preparatoria's contacts attended each activity)")
        print(f"\nPreparatorias:")
        
        # Display with better formatting
        if len(prepa_activity_pct_df) > 0:
            # Filter to only columns that exist in the dataframe
            available_activity_cols = [col for col in top_activities_list[:8] if col in prepa_activity_pct_df.columns]
            display_cols = ["preparatoria", "total_contacts"] + available_activity_cols
            
            if len(display_cols) > 2:  # Make sure we have at least some activity columns
                print(prepa_activity_pct_df[display_cols].to_string(index=False))
            else:
                print("  ‚ö†Ô∏è No activity columns available to display")
                print(prepa_activity_pct_df[["preparatoria", "total_contacts"]].to_string(index=False))

        # ============================================================================
        # Activity Type Participation by Preparatoria
        # ============================================================================
        
        print(f"\n\n  üìä Activity TYPE Participation by Preparatoria")
        print(f"  (Grouped by entry channel type: Digital, Event, Messaging, Niche)")
        
        # Calculate participation by activity TYPE for each prepa
        prepa_activity_type_participation = []
        
        for prepa in top_prepa_list[:20]:
            prepa_contacts = X_with_prepa[X_with_prepa["preparatoria"] == prepa]
            total_prepa_contacts = len(prepa_contacts)
            
            # Count contacts in each activity type
            digital_count = 0
            event_count = 0
            messaging_count = 0
            niche_count = 0
            
            for idx, contact in prepa_contacts.iterrows():
                activities_text = contact.get("apreu_hist_all", "").lower()
                
                # Check which activity types this contact participated in
                if detect_activity_type(activities_text, DIGITAL_ACTIVITIES) > 0:
                    digital_count += 1
                if detect_activity_type(activities_text, EVENT_ACTIVITIES) > 0:
                    event_count += 1
                if detect_activity_type(activities_text, MESSAGING_ACTIVITIES) > 0:
                    messaging_count += 1
                if detect_activity_type(activities_text, NICHE_ACTIVITIES) > 0:
                    niche_count += 1
            
            prepa_activity_type_participation.append({
                "preparatoria": prepa,
                "total_contacts": total_prepa_contacts,
                "Digital_%": round(digital_count / total_prepa_contacts * 100, 1) if total_prepa_contacts > 0 else 0,
                "Event_%": round(event_count / total_prepa_contacts * 100, 1) if total_prepa_contacts > 0 else 0,
                "Messaging_%": round(messaging_count / total_prepa_contacts * 100, 1) if total_prepa_contacts > 0 else 0,
                "Niche_%": round(niche_count / total_prepa_contacts * 100, 1) if total_prepa_contacts > 0 else 0
            })
        
        prepa_activity_type_df = pd.DataFrame(prepa_activity_type_participation)
        prepa_activity_type_df = prepa_activity_type_df.sort_values("total_contacts", ascending=False)
        
        print(f"\n  Preparatorias by Activity Type:")
        print(prepa_activity_type_df.to_string(index=False))

        # ============================================================================
        # Identify Preparatoria Specializations
        # ============================================================================
        
        print(f"\n\n  üéØ Preparatoria Specializations (Dominant Activity Types):")
        
        for idx, row in prepa_activity_type_df.iterrows():
            prepa = row["preparatoria"]
            
            # Find dominant activity type
            activity_types = {
                "Digital": row["Digital_%"],
                "Event": row["Event_%"],
                "Messaging": row["Messaging_%"],
                "Niche": row["Niche_%"]
            }
            
            # Get top 2 activity types
            sorted_types = sorted(activity_types.items(), key=lambda x: x[1], reverse=True)
            top_type = sorted_types[0]
            second_type = sorted_types[1]
            
            if top_type[1] > 0:
                print(f"\n  {prepa}:")
                print(f"    Primary: {top_type[0]} ({top_type[1]}%)")
                if second_type[1] > 0:
                    print(f"    Secondary: {second_type[0]} ({second_type[1]}%)")
                print(f"    Total contacts: {row['total_contacts']}")

print("\n‚úÖ Activity participation by preparatoria analysis complete")

```

```python
# ============================================================================
# Conversion Events & Lifecycle Analysis
# ============================================================================

print("üîÑ Analyzing conversion events and lifecycle stages...")

# ============================================================================
# First Conversion Events
# ============================================================================

first_conv_counts = X_analyzed[X_analyzed["first_conversion"] != ""]["first_conversion"].value_counts().head(20)
first_conv_df = first_conv_counts.to_frame("count")
first_conv_df["percentage"] = (first_conv_df["count"] / first_conv_df["count"].sum() * 100).round(2)
first_conv_df = first_conv_df.reset_index()
first_conv_df.columns = ["first_conversion", "count", "percentage"]

print(f"\n  üìä Top 20 First Conversion Events:")
print(first_conv_df.head(10))

# ============================================================================
# Recent Conversion Events
# ============================================================================

recent_conv_counts = X_analyzed[X_analyzed["recent_conversion"] != ""]["recent_conversion"].value_counts().head(20)
recent_conv_df = recent_conv_counts.to_frame("count")
recent_conv_df["percentage"] = (recent_conv_df["count"] / recent_conv_df["count"].sum() * 100).round(2)
recent_conv_df = recent_conv_df.reset_index()
recent_conv_df.columns = ["recent_conversion", "count", "percentage"]

print(f"\n  üìä Top 20 Recent Conversion Events:")
print(recent_conv_df.head(10))

# ============================================================================
# Conversion Events by Segment
# ============================================================================

conv_by_segment = X_analyzed.groupby("segment_c3").agg({
    "first_conversion": lambda x: x[x != ""].mode()[0] if len(x[x != ""]) > 0 else "None",
    "recent_conversion": lambda x: x[x != ""].mode()[0] if len(x[x != ""]) > 0 else "None",
    "conversion_journey_days": ["mean", "median"]
}).round(2)

conv_by_segment.columns = ['_'.join(col).strip('_') if col[1] else col[0] for col in conv_by_segment.columns.values]
conv_by_segment = conv_by_segment.reset_index()

print(f"\n  üìä Conversion Patterns by Segment:")
print(conv_by_segment)

# ============================================================================
# First Conversion Event Performance
# ============================================================================

# Calculate close rate by first conversion event (for top events)
conv_event_perf = []
for event in first_conv_df.head(15)["first_conversion"]:
    contacts_with_event = X_analyzed[X_analyzed["first_conversion"] == event]
    
    if len(contacts_with_event) > 0:
        close_rate = contacts_with_event["is_closed"].mean() * 100
        avg_days = contacts_with_event["days_to_close"].mean()
        
        conv_event_perf.append({
            "first_conversion": event,
            "total_contacts": len(contacts_with_event),
            "closed_count": contacts_with_event["is_closed"].sum(),
            "close_rate_pct": round(close_rate, 2),
            "avg_days_to_close": round(avg_days, 1) if pd.notna(avg_days) else np.nan
        })

conv_event_perf_df = pd.DataFrame(conv_event_perf).sort_values("close_rate_pct", ascending=False)

print(f"\n  üìä Top Converting First Conversion Events:")
print(conv_event_perf_df.head(10))

# ============================================================================
# Lifecycle Stage Analysis
# ============================================================================

lifecycle_dist = X_analyzed["lifecycle_stage"].value_counts().to_frame("count")
lifecycle_dist["percentage"] = (lifecycle_dist["count"] / lifecycle_dist["count"].sum() * 100).round(2)
lifecycle_dist = lifecycle_dist.reset_index()
lifecycle_dist.columns = ["lifecycle_stage", "count", "percentage"]

print(f"\n  üìä Lifecycle Stage Distribution:")
print(lifecycle_dist)

# Lifecycle by segment
lifecycle_by_segment = pd.crosstab(
    X_analyzed["segment_c3"],
    X_analyzed["lifecycle_stage"],
    normalize='index'
).mul(100).round(2)

print(f"\n  üìä Lifecycle % by Segment:")
print(lifecycle_by_segment)

# ============================================================================
# Time-to-Close Buckets Analysis
# ============================================================================

ttc_dist = X_analyzed["ttc_bucket"].value_counts().to_frame("count")
ttc_dist["percentage"] = (ttc_dist["count"] / ttc_dist["count"].sum() * 100).round(2)
ttc_dist = ttc_dist.reset_index()
ttc_dist.columns = ["ttc_bucket", "count", "percentage"]

print(f"\n  üìä Time-to-Close Bucket Distribution:")
print(ttc_dist)

# TTC by segment
ttc_by_segment = pd.crosstab(
    X_analyzed["segment_c3"],
    X_analyzed["ttc_bucket"],
    normalize='index'
).mul(100).round(2)

print(f"\n  üìä TTC Bucket % by Segment:")
print(ttc_by_segment)

print("\n‚úÖ Conversion events and lifecycle analysis complete")
```

```python
# ============================================================================
# Fast Closers & Slow Closers Analysis
# ============================================================================

print("‚ö° Analyzing fast and slow closers...")

# Define fast closers (‚â§30 days) and slow closers (>120 days)
X_analyzed["is_fast_closer"] = (X_analyzed["days_to_close"] <= 30) & (X_analyzed["days_to_close"].notna())
X_analyzed["is_slow_closer"] = (X_analyzed["days_to_close"] > 120) & (X_analyzed["days_to_close"].notna())

# Fast closers by segment √ó activity
fast_closers_seg_act = []
for segment in X_analyzed["segment_c3"].unique():
    seg_data = X_analyzed[(X_analyzed["segment_c3"] == segment) & (X_analyzed["is_fast_closer"] == True)]
    
    # Get top activities for fast closers in this segment
    all_acts = []
    for activities in seg_data["apreu_activities_list"]:
        if isinstance(activities, list) and activities:
            all_acts.extend(activities)
    
    if all_acts:
        activity_counts = Counter(all_acts)
        for activity, count in activity_counts.most_common(5):
            fast_closers_seg_act.append({
                "segment": segment,
                "activity": activity,
                "fast_closer_count": count
            })

fast_closers_seg_act_df = pd.DataFrame(fast_closers_seg_act)

if not fast_closers_seg_act_df.empty:
    fast_closers_pivot = fast_closers_seg_act_df.pivot_table(
        index="segment",
        columns="activity",
        values="fast_closer_count",
        fill_value=0
    ).astype(int)
    print(f"\n  ‚ö° Fast Closers (‚â§30 days) by Segment √ó Activity:")
    print(fast_closers_pivot.head(10))
else:
    fast_closers_pivot = pd.DataFrame()

# Fast closers by segment √ó preparatoria
fast_closers_seg_prepa = X_analyzed[(X_analyzed["is_fast_closer"] == True) & (X_analyzed["preparatoria"] != "Unknown")]
if len(fast_closers_seg_prepa) > 0:
    fast_closers_prepa = pd.crosstab(
        fast_closers_seg_prepa["segment_c3"],
        fast_closers_seg_prepa["preparatoria"]
    )
    # Keep only top 10 prepas
    top_cols = fast_closers_prepa.sum().sort_values(ascending=False).head(10).index
    fast_closers_prepa = fast_closers_prepa[top_cols]
    print(f"\n  ‚ö° Fast Closers by Segment √ó Top Preparatorias:")
    print(fast_closers_prepa)
else:
    fast_closers_prepa = pd.DataFrame()

# Slow closers by segment √ó activity
slow_closers_seg_act = []
for segment in X_analyzed["segment_c3"].unique():
    seg_data = X_analyzed[(X_analyzed["segment_c3"] == segment) & (X_analyzed["is_slow_closer"] == True)]
    
    # Get top activities for slow closers in this segment
    all_acts = []
    for activities in seg_data["apreu_activities_list"]:
        if isinstance(activities, list) and activities:
            all_acts.extend(activities)
    
    if all_acts:
        activity_counts = Counter(all_acts)
        for activity, count in activity_counts.most_common(5):
            slow_closers_seg_act.append({
                "segment": segment,
                "activity": activity,
                "slow_closer_count": count
            })

slow_closers_seg_act_df = pd.DataFrame(slow_closers_seg_act)

if not slow_closers_seg_act_df.empty:
    slow_closers_pivot = slow_closers_seg_act_df.pivot_table(
        index="segment",
        columns="activity",
        values="slow_closer_count",
        fill_value=0
    ).astype(int)
    print(f"\n  üê¢ Slow Closers (>120 days) by Segment √ó Activity:")
    print(slow_closers_pivot.head(10))
else:
    slow_closers_pivot = pd.DataFrame()

print("\n‚úÖ Fast/slow closers analysis complete")

```

```python
# ============================================================================
# Excel Export - Comprehensive Analysis Workbook
# ============================================================================

print("üìä Creating comprehensive Excel workbook...")

excel_filename = "segments_cluster3_summary.xlsx"

with pd.ExcelWriter(excel_filename, engine='openpyxl') as writer:
    
    # ========================================================================
    # CATEGORY 1: EXECUTIVE SUMMARY
    # ========================================================================
    
    # 1. Executive Summary
    exec_summary = pd.DataFrame({
        "Metric": [
            "Total Contacts Analyzed",
            "Unknown Segment Contacts",
            "3A (Digital-first) Count",
            "3B (Event-first) Count",
            "3C (Messaging-first) Count",
            "3D (Niche) Count",
            "Overall Close Rate %",
            "Avg Days to Close",
            "Total APREU Activities",
            "Unique Activities",
            "Contacts with Preparatoria Data",
            "Unique Preparatorias"
        ],
        "Value": [
            len(X_analyzed),
            (X["segment_c3"] == "Unknown").sum(),
            (X_analyzed["segment_c3"] == "3A (Digital-first)").sum(),
            (X_analyzed["segment_c3"] == "3B (Event-first)").sum(),
            (X_analyzed["segment_c3"] == "3C (Messaging-first)").sum(),
            (X_analyzed["segment_c3"] == "3D (Niche/Low-volume)").sum(),
            round(X_analyzed["is_closed"].mean() * 100, 2),
            round(X_analyzed["days_to_close"].mean(), 1),
            len(all_activities),
            len(activity_counts),
            len(X_with_prepa),
            X_with_prepa["preparatoria"].nunique()
        ]
    })
    exec_summary.to_excel(writer, sheet_name="1_executive_summary", index=False)
    
    # ========================================================================
    # CATEGORY 2: ENTRY CHANNEL SEGMENTS (3A-3D)
    # ========================================================================
    
    # 2. Segment Counts
    segment_counts.to_excel(writer, sheet_name="2_segment_counts", index=False)
    
    # 3. Segment Distribution (same as counts, formatted differently)
    seg_dist = segment_counts.copy()
    seg_dist.to_excel(writer, sheet_name="3_segment_distribution", index=False)
    
    # 4. Segment Performance (create if not exists)
    seg_perf = X_analyzed.groupby("segment_c3").agg({
        "contact_id": "count",
        "is_closed": ["sum", "mean"],
        "days_to_close": "mean",
        "engagement_score": "mean",
        "apreu_activity_count": "mean",
        "likelihood_pct": "mean"
    })
    seg_perf.columns = ['_'.join(col).strip('_') if col[1] else col[0] for col in seg_perf.columns.values]
    seg_perf = seg_perf.reset_index()
    seg_perf = seg_perf.rename(columns={
        "contact_id_count": "total_contacts",  # FIXED: Column name is contact_id_count after aggregation
        "is_closed_sum": "closed_count",
        "is_closed_mean": "close_rate",
        "days_to_close_mean": "avg_days_to_close",
        "engagement_score_mean": "avg_engagement",
        "apreu_activity_count_mean": "avg_activities",
        "likelihood_pct_mean": "avg_likelihood_pct"
    })
    seg_perf["close_rate_pct"] = (seg_perf["close_rate"] * 100).round(2)
    seg_perf["avg_days_to_close"] = seg_perf["avg_days_to_close"].round(1)
    seg_perf["avg_engagement"] = seg_perf["avg_engagement"].round(2)
    seg_perf["avg_activities"] = seg_perf["avg_activities"].round(1)
    seg_perf["avg_likelihood_pct"] = seg_perf["avg_likelihood_pct"].round(1)
    
    
    # 4. Segment Performance
    seg_perf.to_excel(writer, sheet_name="4_segment_performance", index=False)
    
    # ========================================================================
    # CATEGORY 3: APREU ACTIVITIES ANALYSIS
    # ========================================================================
    
    # 5. Activity Participation
    activity_df.to_excel(writer, sheet_name="5_activity_participation", index=False)
    
    # 6. Activity by Segment
    if not activity_by_segment.empty:
        activity_by_segment.to_excel(writer, sheet_name="6_activity_by_segment")
    
    # 7. Activity Diversity Distribution
    activity_diversity_dist = X_analyzed["apreu_activity_diversity"].value_counts().sort_index().to_frame("count")
    activity_diversity_dist.to_excel(writer, sheet_name="7_activity_diversity")
    
    # 8. Top Activities Overall
    activity_df.head(20).to_excel(writer, sheet_name="8_top_activities_overall", index=False)
    
    # 9. Activity Conversion Rates
    activity_conversion_df.to_excel(writer, sheet_name="9_activity_conversion_rates", index=False)
    
    # ========================================================================
    # CATEGORY 4: PREPARATORIA ANALYSIS
    # ========================================================================
    
    # 10. Top Prepas Overall
    top_prepas.to_excel(writer, sheet_name="10_top_prepas_overall", index=False)
    
    # 11. Prepas by Segment
    prepas_by_segment.to_excel(writer, sheet_name="11_prepas_by_segment")
    
    # 12. Prepas by Activity
    if not prepas_by_activity.empty:
        prepas_by_activity.to_excel(writer, sheet_name="12_prepas_by_activity")
    
    # 13. Prepa Performance
    prepa_perf.to_excel(writer, sheet_name="13_prepa_performance", index=False)
    
    # 14. Prepa Year Distribution
    prep_year_dist.to_excel(writer, sheet_name="14_prepa_year_distribution", index=False)
    
    # 14a. Activity Participation % by Preparatoria (NEW)
    if not prepa_activity_pct_df.empty:
        prepa_activity_pct_df.to_excel(writer, sheet_name="14a_prepa_activity_pct", index=False)
    
    # 14b. Activity Type Participation by Preparatoria (NEW)
    if not prepa_activity_type_df.empty:
        prepa_activity_type_df.to_excel(writer, sheet_name="14b_prepa_activity_type", index=False)
    
    # ========================================================================
    # CATEGORY 5: CONVERSION EVENT ANALYSIS
    # ========================================================================
    
    # 15. First Conversion Events
    first_conv_df.to_excel(writer, sheet_name="15_first_conversion_events", index=False)
    
    # 16. Recent Conversion Events
    recent_conv_df.to_excel(writer, sheet_name="16_recent_conversion_events", index=False)
    
    # 17. Conversion by Segment
    conv_by_segment.to_excel(writer, sheet_name="17_conversion_by_segment", index=False)
    
    # 18. Conversion Timeline (journey days distribution)
    conv_timeline = X_analyzed[X_analyzed["conversion_journey_days"].notna()]["conversion_journey_days"].describe().to_frame("days")
    conv_timeline.to_excel(writer, sheet_name="18_conversion_timeline")
    
    # 19. Conversion Event Performance
    conv_event_perf_df.to_excel(writer, sheet_name="19_conversion_event_performance", index=False)
    
    # ========================================================================
    # CATEGORY 6: ENGAGEMENT & EMAIL METRICS
    # ========================================================================
    
    # 20. Engagement by Segment
    engagement_by_seg = X_analyzed.groupby("segment_c3")[["num_sessions", "num_pageviews", "forms_submitted", "engagement_score"]].mean().round(2)
    engagement_by_seg.to_excel(writer, sheet_name="20_engagement_by_segment")
    
    # 21. Email Metrics by Segment
    email_by_seg = X_analyzed.groupby("segment_c3")[["email_delivered", "email_opened", "email_clicked", "email_engagement_score"]].mean().round(2)
    email_by_seg.to_excel(writer, sheet_name="21_email_metrics_by_segment")
    
    # 22. Email Engagement Score Distribution
    email_score_dist = X_analyzed["email_engagement_score"].describe().to_frame("email_engagement_score")
    email_score_dist.to_excel(writer, sheet_name="22_email_engagement_score")
    
    # 23. Digital Efficiency
    digital_eff = X_analyzed.groupby("segment_c3")[["pageviews_per_session", "forms_per_session"]].mean().round(3)
    digital_eff.to_excel(writer, sheet_name="23_digital_efficiency")
    
    # ========================================================================
    # CATEGORY 7: BUSINESS OUTCOMES & LIFECYCLE
    # ========================================================================
    
    # 24. Likelihood by Segment
    likelihood_by_seg = X_analyzed.groupby("segment_c3")["likelihood_pct"].describe().round(2)
    likelihood_by_seg.to_excel(writer, sheet_name="24_likelihood_by_segment")
    
    # 25. Closure Stats by Segment
    closure_stats = seg_perf[["segment_c3", "total_contacts", "closed_count", "close_rate_pct", 
                               "avg_days_to_close"]]
    closure_stats.to_excel(writer, sheet_name="25_closure_stats_by_segment", index=False)
    
    # 26. Lifecycle by Segment
    lifecycle_by_segment.to_excel(writer, sheet_name="26_lifecycle_by_segment")
    
    # 27. TTC Buckets by Segment
    ttc_by_segment.to_excel(writer, sheet_name="27_ttc_buckets_by_segment")
    
    # ========================================================================
    # CATEGORY 8: TIME-TO-CLOSE DEEP DIVE
    # ========================================================================
    
    # 28. Comprehensive TTC by Segment
    ttc_comprehensive = X_analyzed.groupby(["segment_c3", "ttc_bucket"]).agg({
        "contact_id": "count",
        "days_to_close": ["mean", "median"]
    }).round(2)
    ttc_comprehensive.columns = ['_'.join(col).strip('_') for col in ttc_comprehensive.columns.values]
    ttc_comprehensive = ttc_comprehensive.reset_index()
    ttc_comprehensive.to_excel(writer, sheet_name="28_comprehensive_ttc_segment", index=False)
    
    # 29. Fast Closers Segment √ó Activity
    if not fast_closers_pivot.empty:
        fast_closers_pivot.to_excel(writer, sheet_name="29_fast_closers_seg_x_act")
    
    # 30. Slow Closers Segment √ó Activity
    if not slow_closers_pivot.empty:
        slow_closers_pivot.to_excel(writer, sheet_name="30_slow_closers_seg_x_act")
    
    # 31. Fast Closers Segment √ó Prepa
    if not fast_closers_prepa.empty:
        fast_closers_prepa.to_excel(writer, sheet_name="31_fast_closers_seg_x_prepa")
    
    # 32. Overall TTC Summary
    ttc_dist.to_excel(writer, sheet_name="32_overall_ttc_summary", index=False)
    
    # ========================================================================
    # METADATA
    # ========================================================================
    
    # Metadata
    metadata = pd.DataFrame({
        "Field": [
            "Analysis Timestamp",
            "Total Contacts in Dataset",
            "Analyzed Contacts",
            "Unknown Segment Count",
            "Cluster Type",
            "Focus",
            "Entry Channel Segments"
        ],
        "Value": [
            datetime.now().strftime("%Y-%m-%d %H:%M:%S"),
            len(X),
            len(X_analyzed),
            (X["segment_c3"] == "Unknown").sum(),
            "Cluster 3",
            "Promotion-driven Converters (APREU Activities)",
            "3A (Digital), 3B (Event), 3C (Messaging), 3D (Niche)"
        ]
    })
    metadata.to_excel(writer, sheet_name="metadata", index=False)

print(f"‚úÖ Excel workbook created: {excel_filename}")
print(f"   üìä Contains 32+ worksheets with comprehensive analysis")



```
